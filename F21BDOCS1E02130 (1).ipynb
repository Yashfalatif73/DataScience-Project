```python
categorical=data[["satisfaction","Customer Type","Type of Travel","Class"]]
columns = categorical.columns

# Set up the figure and axis
fig, axes = plt.subplots(len(columns), 1, figsize=(10, 5*len(columns)))

# Loop through each categorical column and plot a bar plot
for i, column in enumerate(columns):
    ax = axes[i] if len(columns) > 1 else axes
    value_counts = categorical[column].value_counts()
    value_counts.plot(kind='bar', ax=ax, color='skyblue', edgecolor='black')
    ax.set_title(column)
    ax.set_xlabel('Category')
    ax.set_ylabel('Count')
    ax.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust layout
plt.tight_layout()
plt.show()
```
